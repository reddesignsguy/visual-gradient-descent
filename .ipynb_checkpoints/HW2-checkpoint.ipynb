### 1a) H is symmetric

A matrix $H$ is symmetric if $H = H^T$ (the transpose)

Given $H = X(X^TX)^{-1}X^T$, prove that $H = H^T$ or, rather, $X(X^TX)^{-1}X^T = (X(X^TX)^{-1}X^T)^T$ 

#### Prove $H^T = H$:

$$H^T = \left(X(X^TX)^{-1}X^T\right)^T$$

1. Distribute:

$$H^T = X^T\left((X^TX)^{-1}\right)^T\left(X^T\right)^T$$

2. A double transpose of matrix X gives us X
3. The transpose of an inversed matrix is the inverse of the transposed matrix

$$H^T = X\left((X^TX)^T\right)^{-1}X^T$$

4. Distribute

$$H^T = X(X^TX)^{-1}X^T$$

5. Final

$$H^T = X(X^TX)^{-1}X^T = H$$

Therefore, $H$ is symmetric as $H^T = H$ assuming that $X^TX$ is invertible

### 1b) H is can still be symmetric even if $X^TX$ is not invertible

* Using the property that Pseudo-inverse of a tranposed matrix is = to the transpose of the pseudo-inverse of a matrix, H is still symmetric even if $X^TX$ is not invertible.

1. <b> Same process as above except using the pseudo inverse </b>

  $$H^T = \left(X(X^TX)^{+}X^T\right)^T$$

$$H^T = X^T\left((X^TX)^{+}\right)^T\left(X^T\right)^T$$

2. The transpose of an PSEUDO-inversed matrix is the PSEUDO-inverse of the transposed matrix

$$H^T = X\left((X^TX)^T\right)^{+}X^T$$

$$H^T = X(X^TX)^{+}X^T$$

$$H^T = X(X^TX)^{+}X^T = H$$

Therefore, $X^TX$ does not have to be invertible for $H$ to be symmetric

### 2) Artificial Dataset and Applying Regression

```python
# -=-=-=-=-=-=-=-=-=- part a: f(x) -=-=-=-=-=-=-=-=-=-=-=
# 0) Target function
def f(x):
    return x**3 + 2*x**2 - 5*x - 6

# 1) Generate data 
x_values = np.random.uniform(-4, 4, 30)

# 2) Calculate y values 
y_values = f(x_values) + np.random.uniform(-2, 2, 30)

# 3) Convert data 
data = pd.DataFrame({'x': x_values, 'y': y_values})

# 4) Visualize it
plt.figure(figsize=(10, 6))
plt.scatter(data['x'], data['y'], color='#123123') # data

x_range = np.linspace(-4, 4) 
plt.plot(x_range, f(x_range), color='red', label='f(x) = x^3 + 2x^2 - 5x - 6') # f(x)

plt.title('Linear Regression Demonstration')
plt.xlabel('x')
plt.ylabel('y')

# -=-=-=-=-=-=-==-=-=-=-==-=-= part b: g(x) -=-=-=-=-=-=-=-=-=-=-=-=
# 1) Create X out of the data points
getRowVector = lambda x: [1, x]
X = np.array([getRowVector(pt) for pt in x_values])

# 2) Get Pseudo inverse of X
X_PI = np.linalg.pinv(X)

# 3) Calculate weights
weights = np.matmul(X_PI, y_values)

# 4) Create g(x)
def g(x):
    w = weights
    return w[0] + w[1]*x

# 5) Visualize it
plt.plot(x_range, g(x_range), color = "#A5E3FF", label=f'g(x) = {round(weights[1],2)}x + {round(weights[0], 2)}')


# -=-=-=-=-=-=-=-=-=-=-=-= part c: g(x) as a third order polynomial -=-=-=-=-=-=-=-=
# 1) Create X out of the data points
getRowVector = lambda x: [1, x, x**2, x**3]
X = np.array([getRowVector(pt) for pt in x_values])

# 2) Get Pseudo inverse of X
X_PI = np.linalg.pinv(X)

# 3) Calculate weights
weights = np.matmul(X_PI, y_values)

# 4) Create g(x)
def g(x):
    w = weights
    return w[0] + w[1]*x + w[2]*x**2 + w[3]*x**3

print(weights)

# 5) Visualize it
plt.plot(x_range, g(x_range), color = "#1BB7FF", label=f'3rd order g(x) = {round(weights[3],2)}x^3 + {round(weights[2],1)}x^2 + {round(weights[1],2)}x + {round(weights[0], 2)}')

# -=-=-=-=-=-= Show Plot -=-=-=-=-=-=-=-=
plt.legend()
plt.grid()
plt.show()
```

### Differences
- The 3rd polynomial function is much more accurate because there are more features that describe the data. Also, we already know f(x) is a 3rd polynomial function, which gets us very close to f(x) compared to other functions of other degrees
### Similarities
- Both functions try to match the data points and minimize the error for each pt. This is because we used the linear regression algorithm which includes calculating the pseudo-inverse of X.

```python
# -=-=-=-=-=-=-=-=-=-=-=-=-= Create Model -=-=-=-=-=-=-=-=-=-=-=-=
# Function that graphs all hypotheses of the model
def graph_hypotheses(model, end_range = 10):
    for i in range(len(model.history)):
        xs = np.linspace(0, end_range)
        
        if model.history[i][2] != 0:
            ys = -model.history[i][1]/model.history[i][2] * xs - model.history[i][0]/model.history[i][2]
            
            
            if i == len(model.history) - 1:
                plt.plot(xs,ys, linewidth=(0.8), alpha = (i/len(model.history)), color="black", label=f"Final {len(model.history)}th hypothesis: g(x)")
                print("Final hypothesis: ", model.history[i])
            else:
                plt.plot(xs,ys, linewidth=(0.1), alpha = ((i)/len(model.history)), color="#D8D7D8")

class model:
    def __init__(self, w):
        self.w = w
        self.updates = 0
        self.history = []
        self.history.append(w.copy())
    
    def h(self, x):
        return np.sign(self.w[0] + self.w[1]*x[1] + self.w[2]*x[2])
    
    def h_custom_weights(self, x, w):
        return np.sign(w[0] + w[1]*x[1] + w[2]*x[2])
    
    def updateWeights(self, x):
        self.w[0] = self.w[0] + y(x)*x[0]
        self.w[1] = self.w[1] + y(x)*x[1]
        self.w[2] = self.w[2] + y(x)*x[2]
        
        self.updates += 1
        
        self.history.append(self.w.copy())

# A better model class :D
class model_2(model):
        def __init__(self, w):
            super().__init__(w)
            self.updates = 0
            self.history = []
            self.history.append(w.copy())
            
        def updateWeights(self, x, y):
            self.w[0] = self.w[0] + y*x[0]
            self.w[1] = self.w[1] + y*x[1]
            self.w[2] = self.w[2] + y*x[2]

            self.updates += 1

            self.history.append(self.w.copy())
        
        # This function isn't used for data set A, but is used for all other data sets
        def trainModel(self, data, n = 10000):
            # n limits the amt of iterations or updates

            # Run PLA until there are no misclassified points
            
            while n > 0:
                no_msclssified = True
                for i in range(len(data)):
                    correct_classification = data[i][2]
                    # Vectorize data
                    x = [1, data[i][0], data[i][1]]

                    # Misclassified point
                    if self.h(x) != correct_classification:
                        if correct_classification == self.h(x):
                            print("We are saying something is misclassified when it isn't")
                            
                        self.updateWeights(x, data[i][2])
                        no_msclssified = False
                        n-=1


                if no_msclssified:
                    break

# -=-=-=-=-=-=-=-=-= Define symmetry and intensity -=-=-=-=-=-=-=-=
import matplotlib.patches as mpatches

def getSymmetryRating(values):
    sideLen = len(values)
    
    symmetryRating = 0
    for r in range(0, sideLen):
        for c in range(0, int(sideLen/2)):
            pixel_value = values[r][c]
            complementary_pixel_value = values[r][sideLen - c - 1]
            
            # Allow some tolerance for how different they can be
            if abs(pixel_value - complementary_pixel_value) < 0.01:
                symmetryRating+=1
    
    return symmetryRating

def getIntensityRating(values):
    sideLen = len(values)
    
    intensityRating = 0
    for r in range(0, sideLen):
        for c in range(0, int(sideLen/3)):
            pixel_value = values[r][c]
            
            if pixel_value > -0.9:
                intensityRating += 1

    for r in range(0, sideLen):
        for c in range(int(sideLen*2/3), int(sideLen)):
            pixel_value = values[r][c]
            
            if pixel_value > -0.9:
                intensityRating += 1
                
    return intensityRating

processedZipDigits = open("processedZipDigits.train", 'r')


# -=-=-=-=-= Go through every data point and extract its intensity and symmetry features -=-=-=-=-=-=
featureData = []

for line in processedZipDigits:
    line_array = line.strip().split(' ')
    classification = float(line_array[0])
    pixel_values = line_array[1:]
    
    # Convert strings to float and reshape into a 16x16 array
    pixel_values = np.array(pixel_values, dtype=float).reshape(16, 16)

    symmetry_rating = getSymmetryRating(pixel_values)
    intensity_rating = getIntensityRating(pixel_values)
    featureData.append((symmetry_rating, intensity_rating, classification))

# -=-=-=-=-=-= Visualize the data points relative to their ratings -=-=-=-=-=-=-=
def scatterPoints():
    oneLabelPlaced = False
    negativeLabelPlaced = False
    for i, pt in enumerate(featureData):
        symmetryRating = pt[0]
        intensityRating = pt[1]
        classification = pt[2]
        
        if classification == 1:
            plt.scatter(symmetryRating, intensityRating, color = '#49CBDB', label="1" if not oneLabelPlaced else None)
            oneLabelPlaced = True
        else:
            plt.scatter(symmetryRating, intensityRating, color = "#DA3030", label="-1" if not negativeLabelPlaced else None)
            negativeLabelPlaced = True

scatterPoints()

# -=-=-=-=-=-=-=-=-= New g(x) graphing function -=-=-=-=-=-=-=-=
def graph_hypotheses(model, end_range = 10):
    for i in range(len(model.history)):
        xs = np.linspace(0, end_range)
        
        if model.history[i][2] != 0:
            ys = -model.history[i][1]/model.history[i][2] * xs - model.history[i][0]/model.history[i][2]
            
            
            if i == len(model.history) - 1:
                plt.plot(xs,ys, linewidth=(0.8), alpha = (i/len(model.history)), color="black", label=f"Final 1000th hypothesis: g(x)")
                print("Final hypothesis: ", model.history[i])
            else:
                plt.plot(xs,ys, linewidth=(0.1), alpha = ((i)/len(model.history)), color="#D8D7D8")
# -=-=-=-=-=-=-=-= Perform PLA -=-=-=-=-=-=-=-=-=-=

# Calculuate a relatively close initial set of weights just by looking at the data
# 1. w = -(-25/80) * x - (240/80)
# 2. w = 25/80 * x - 3

initialWeights = [300,-25,80]
modelDigits = model_2(initialWeights)
modelDigits.trainModel(featureData, n=1000)

graph_hypotheses(modelDigits, 128)


plt.xlim(0,130)
plt.ylim(0,130)
plt.title("Is the number a 1? PLA (no modifications)")
plt.ylabel("Intensity")
plt.xlabel("Symmetry")
plt.legend()
plt.show()

# -=-=-=-=-=-=-=-=- Pocket Algorithm: Calculate the error for each hypothesis and find the iteration w/ least error -=-=-=-=-=-
best_hypothesis = None
best_error = None
best_error_iteration = None

def linearized_h(x, w0, w1, w2):
    return -(w1/w2)*x - (w0/w2)


for i, h in list(enumerate(modelDigits.history)):
    w0 = h[0]
    w1 = h[1]
    w2 = h[2]
    
    if w2 == 0:
        continue
        
    error = 0
    num_misclassified = 0
    for data_pt in featureData:
        x = data_pt[0]
        y = data_pt[1] 
        correct_classification = data_pt[2]
        
            
        guess = linearized_h(x, w0,w1,w2)
        binarized_guess = 1 if y < guess else -1
        # Find a misclassified point
        if correct_classification != binarized_guess:
            error += 1
            
    if not best_error or error < best_error:
        best_error = error
        best_hypothesis = [w0,w1,w2]
        best_error_iteration = i
    
        
    
    plt.scatter(i, error, color="red")

    
plt.scatter(best_error_iteration, best_error, color="blue", label = "H(x) at iteration {best_error_iteration} minimizes error")

print(f"Best error of {best_error} for weights: {best_hypothesis} at iteration {best_error_iteration}")

plt.legend();
plt.xlabel("Iteration i")
plt.ylabel("Error")
plt.title("Error vs Iteration")
plt.show()


# -=-=-=-=-=-=- Regraph the data but with the algorithmically found best hypothesis (Pocket Algorithm) -=-=-=-=-=-=-

scatterPoints()
plt.title("Is the number a 1? (PLA + Pocket Algorithm)")
plt.ylabel("Intensity")
plt.xlabel("Symmetry")
plt.xlim(0,130)
plt.ylim(0,130)

x = np.linspace(0,130)
plt.plot(x, -(best_hypothesis[1]/best_hypothesis[2]) * x - (best_hypothesis[0]/best_hypothesis[2]), color="black", label=f"g(x) - iteration {best_error_iteration}", linewidth=(1.2))
# plt.plot(x, 25/80 * x - 3, color="black", label="some approximate f(x)", linestyle='dashed', linewidth=(1.2))

plt.legend()
plt.show()

processedZipDigits.close()
```

### 3c) Third-order feature polynomial transformation
#### I. Error Graph

```python
processedZipDigits = open("processedZipDigits.train", 'r')

# a model with 3rd polynomial featue transformation
class model_3(model_2):
        def __init__(self, w):
            super().__init__(w)
            self.updates = 0
            self.history = []
            self.history.append(w.copy())

        def h(self, x, w = None):
            if w is None:
                w = self.w
                
            return np.sign(w[0] + w[1]*x[1] + w[2]*x[2] + w[3]*x[3] + w[4]*x[4] + w[5]*x[5] + w[6]*x[6] + w[7]*x[7] + w[8]*x[8] + w[9]*x[9])

        # This solves for x1 and returns the result. x1 should represent the intensity.
        def x1(self, x0, x1, x2):
            w = self.w
            return (-(w[0]+ w[2]*x[2] + w[3]*x[3] + w[4]*x[4] + w[5]*x[5] + w[6]*x[6] + w[7]*x[7] + w[8]*x[8] + w[9]*x[9])) / w[1]
            
            
        def updateWeights(self, x, y):
            self.w[0] = self.w[0] + y*x[0]
            self.w[1] = self.w[1] + y*x[1]
            self.w[2] = self.w[2] + y*x[2] 
            self.w[3] = self.w[3] + y*x[3]
            self.w[4] = self.w[4] + y*x[4]
            self.w[5] = self.w[5] + y*x[5]
            self.w[6] = self.w[6] + y*x[6]
            self.w[7] = self.w[7] + y*x[7]
            self.w[8] = self.w[8] + y*x[8]
            self.w[9] = self.w[9] + y*x[9]
            

            self.updates += 1

            self.history.append(self.w.copy())
        
        def trainModel(self, data, n = 10000): # n limits the amt of iterations or updates
            # Run PLA until there are no misclassified points
            while n > 0:
                no_msclssified = True
                for i in range(len(data)):
                    sample = data[i]
                    correct_classification = sample[10]
                    
                    # Vectorize data
                    x = [sample[0], sample[1], sample[2], sample[3], sample[4], sample[5], sample[6], sample[7], sample[8], sample[9]]

                    # Misclassified point
                    if self.h(x) != correct_classification:
                        self.updateWeights(x, correct_classification)
                        no_msclssified = False
                        n-=1

                if no_msclssified:
                    break
                    

# -=-=-=-=-= Go through every data point and extract its intensity and symmetry features -=-=-=-=-=-=
featureData = []

for line in processedZipDigits:
    line_array = line.strip().split(' ')
    classification = float(line_array[0])
    pixel_values = line_array[1:]
    
    # Convert strings to float and reshape into a 16x16 array
    pixel_values = np.array(pixel_values, dtype=float).reshape(16, 16)

    symmetry_rating = getSymmetryRating(pixel_values)
    intensity_rating = getIntensityRating(pixel_values)
    featureData.append((intensity_rating, symmetry_rating, classification))

# -=-=-=-=-=-=-== Extract polynomial features from intensity/symmetry -=-=-=-=-=-=-=-=
# NOTE: This returns a vector of 3rd degree polynomial features PLUS the classification of the sample
def third_degree_polynomial(data):
    transformed_data = []
    for x in data:
        x1 = x[0] # feature 1 (intensity)
        x2 = x[1] # feature 2 (symmetry)
        classification = x[2]
        transformed_sample = [1, x1, x2, x1**2, x1*x2, x2**2, x1**3, x2*(x1**2), (x2**2)*x1, x2**3, classification]
        
        transformed_data.append(transformed_sample)
        
    return transformed_data

featureData = third_degree_polynomial(featureData)

# -=-=-=-=-=-= Visualize the data points relative to their ratings -=-=-=-=-=-=-=
def scatterPoints2():
    oneLabelPlaced = False
    negativeLabelPlaced = False
    for i, pt in enumerate(featureData):
        intensityRating = pt[1]  
        symmetryRating = pt[2]   
        classification = pt[10]
        
        if classification == 1:
            plt.scatter(symmetryRating, intensityRating, color = '#49CBDB', label="1" if not oneLabelPlaced else None)
            oneLabelPlaced = True
        else:
            plt.scatter(symmetryRating, intensityRating, color = "#DA3030", label="-1" if not negativeLabelPlaced else None)
            negativeLabelPlaced = True

scatterPoints2()

# -=-=-=-=-=-=-=-= Perform PLA -=-=-=-=-=-=-=-=-=-=
initialWeights = [123,456,789,10,5,6,7,8,9,10]


modelDigits = model_3(initialWeights)
modelDigits.trainModel(featureData, n=1000)

x_range = np.linspace(0, 130) 
# plt.plot(x_range, modelDigits.x1(x_range), color = "#1BB7FF")
plt.xlim(0,130)
plt.ylim(0,130)
plt.title("Is the number a 1?")
plt.ylabel("Intensity")
plt.xlabel("Symmetry")
plt.show()

# -=-=-=-=-=-=-=-=- Pocket Algorithm: Calculate the error for each hypothesis and find the iteration w/ least error -=-=-=-=-=-
best_hypothesis = None
best_error = None
best_error_iteration = None
    
for i, h in list(enumerate(modelDigits.history)):
    w0 = h[0]
    w1 = h[1]
    w2 = h[2]
    w3 = h[3]
    w4 = h[4]
    w5 = h[5]
    w6 = h[6]
    w7 = h[7]
    w8 = h[8]
    w9 = h[9]
    
    if w2 == 0:
        continue
        
    error = 0
    num_misclassified = 0
    for pt in featureData:
        x0 = pt[0]
        x1 = pt[1]
        x2 = pt[2]
        x3 = pt[3]
        x4 = pt[4]
        x5 = pt[5]
        x6 = pt[6]
        x7 = pt[7]
        x8 = pt[8]
        x9 = pt[9]
        correct_classification = pt[10]

        # Plug in values into this hypothesis
        guess = modelDigits.h([x0,x1,x2,x3,x4,x5,x6,x7,x8,x9], [w0,w1,w2,w3,w4,w5,w6,w7,w8,w9])
        
        # If this hypothesis misclassifies this point, then increase error
        if correct_classification != guess:
            error += 1

    # print(f"The weights {h} has an error of: {error}")
        
    # Keep track of the iteration w/ best error
    if not best_error or error < best_error:
        best_error = error
        best_hypothesis = [w0,w1,w2,w3,w4,w5,w6,w7,w8,w9]
        best_error_iteration = i
    
    plt.scatter(i, error, color="red")

print(best_error)
    
plt.scatter(best_error_iteration, best_error, color="blue", label = f"H(x) at iteration {best_error_iteration} minimizes error")

print(f"Best error of {best_error} for weights: {best_hypothesis} at iteration {best_error_iteration}")

plt.legend();
plt.xlabel("Iteration i")
plt.ylabel("Error")
plt.title("Error vs Iteration")
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
 
#  -=-=-=-=-=-=-=-= Generate dataset using sklearn -=-=-=-=-=-=-=-=-=-=-=-=
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, random_state=10, hypercube=False,
                           class_sep=7)

#  -=-=-=-=-=-=-=-=-=-=- Visualize data -=-=-=-=-=-=-=-=-=-=-=-=-=-=
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Random Dataset')
plt.show()

print(X.shape[1])
#  -=-=-=-=-=-=-=--=-=-=-=-= Define gradient descent algorithm -==-=-=-=-=-=-=-=-=-=-=-=-=
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# -=-=-=-=-=-=-=-=-=-=-=-= First version of my error function -=-=-=-=-=-=-=-=-=-=
# def cross_entropy_error(X, y, weights):
#     total_error = 0
    
#     # Iterate thru each sample, calculate the error 
#     for n, sample_x in enumerate(X):
#         # Calculate wTx
#         transposed_weights = weights.copy()
#         transposed_weights.shape = (len(weights), 1)
#         wTx = sample_x.dot(transposed_weights) # row first, then column

#         # Calculate error for this sample
#         e = np.e
#         sample_error = np.log(1 + e**(-(y[n])*wTx))

#         # Add to total error
#         total_error += sample_error
        
#     return total_error / X.shape[0]

# -=-=-=-=-=-=-=-=-=-=-=-=-= Modified using ChatGPT to incorporate gradient function (see sources cited) -=-=-=-=-=-=-=-=-=-=
def cross_entropy_error(X, y, weights):
  num_samples = X.shape[0]

  # Initialize total error
  total_error = 0

  # Initialize gradient (same shape as weights)
  gradient = np.zeros(weights.shape)

  # Iterate through each sample
  for n, sample_x in enumerate(X):
    # Calculate wTx
    transposed_weights = weights.copy()
    transposed_weights.shape = (len(weights), 1)
    wTx = sample_x.dot(transposed_weights)  # row first, then column

    # Calculate predicted probability (a) for this sample (Used for calculating gradient)
    a = 1 / (1 + np.exp(-wTx))

    # Calculate error for this sample
    e = np.e
    sample_error = np.log(1 + e**(-(y[n])*wTx))

    # Add to total error
    total_error += sample_error

    # Update gradient for this sample
    gradient += (a - y[n]) * sample_x

  # Average the total error
  error = total_error / num_samples
  gradient = gradient / num_samples

  return error, gradient

def gradient_descent(X, y, iterations, learning_rate):
    m, n = X.shape # m = num samples, n = num features
    weights = np.array([0.] * n) # initialize weights as 0 for each feature
    
    for i in range(iterations):
        # -=-=-=-=-=-=-=- Vectorized method of descent from chatGPT -=-=-=-=-=-=-=-=
        # z = np.dot(X, weights) # wTx
        # predictions = sigmoid(z) # Plug in wTx
        # error = predictions - y # Error defined as difference between predictions and y values
        # gradient = np.dot(X.T, error) / m 

        # -=-=-=-=-=-=-=-=- Iterative method of descent -=-=-=-=-=-=-=-=-=-=
        error, gradient = cross_entropy_error(X,y,weights)
        weights -= learning_rate * gradient

        if i < 50:
            print(f"Iteration {i} -- Error: {error} , Gradient Function: {gradient}")
        # print(f"Iterative: {cross_entropy_error(X=X, y=y, weights=weights)}")
        # print(f"Vectorized: {cross_entropy_error2(X=X, y=y, weights=weights)}")
    
    return weights
    
# -=-=-=-=-=-=-= Partition data into training and test data -=-=-=-=-=-=-=-=-=-=
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# -=-=-=-=-=-=- Apply gradient descent (4000 iterations, with a learning rate of 0.5) -=-=-=-=--=-=
weights = gradient_descent(X_train, y_train, iterations=4000, learning_rate=0.5)

# -=-=-=-=-=-=-=-=-= Predict using trained model -=-=-==-=-=-=-=-=-
predictions = sigmoid(np.dot(X_test, weights)) # apply xTw (same as wTx) w/ the new weights
predicted_labels = np.round(predictions)

# -=-=-==-=-=-=-=-=- Calculate accuracy -=-=-=-=-==-=-=-=-=-=
accuracy = np.mean(predicted_labels == y_test) # check how many labels are correct
print("Accuracy on out-of-sample data:", accuracy)

# -=-=-=-=-=-= Visualize data -=-=-=-=-=-=-=-=-=-=
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.coolwarm, s=50, edgecolor="k", label="Actual")
plt.scatter(X_test[:, 0], X_test[:, 1], c=predicted_labels, cmap=plt.cm.coolwarm, s=50, marker="x", label="Predicted")
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Test Data with True Labels')
plt.legend()
plt.show()
```

For the above, the CIRCLES are the y data set (answers). The X's are the predictions.

For the gradient, I had to do some research to find equations that matched with what was on the slides. I found out that the gradient could be calculated by calculating the difference between real and predicted sigmoid function, multiplying it by the x sample row vector, then adding this to a running total. I corroborated this process across sources like StackOverFlow and ChatGPT. This process would be repeated, then averaged by dividing it by the number of samples, which is similiar to getting the cross entropy error of summation.

As can seen by the results, this is true.
